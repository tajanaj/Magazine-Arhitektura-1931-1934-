```python
from collections import Counter
import spacy
import textacy
```

```python
nlp = spacy.load("hr_core_news_lg")
```

```python
with open ("kregar.txt", mode='r', encoding='utf-8') as f:
    text_master = f.read().replace("\n", " ")
```

```python
doc = nlp(text_master)
```

```python
parovi_prvi = []
parovi_drugi = []
i = 1

for token in doc:
    if not token.is_punct and not token.is_space:
        if (i % 2) == 0: 
            parovi_drugi.append(token.text)
        else: 
            parovi_prvi.append(token.text)
        i = i+1
        
print(i)
```

```python
print(parovi_prvi)
print(parovi_drugi)
```

```python
parovi = list(zip(parovi_prvi, parovi_drugi))
```

```python
print(parovi)
```

```python
import os
directory = "Rado Kregar"
texts = []

for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        f = os.path.join(directory, filename)
        texts.append(f)
```

```python
pairs = []

for text in texts:
    with open (text, mode='r', encoding='utf-8') as f:
        txt = f.read().replace("\n", " ")
    
    doc = nlp(txt)
    
    sentences = list(doc.sents)
    

    
    text_pairs_sent = []
    
    for sentence in sentences:
    
        lemma_words = []
        for token in sentence:
            if not token.is_stop and not token.is_punct and not token.is_space: 
                lemma_words.append(token.lemma_)

        text_pairs = []

        for par in parovi:
            prvi = par[0]
            drugi = par[1]
            if prvi in lemma_words:
                if drugi in lemma_words:
                    text_pairs.append(par)

        #print(text_pairs)
        text_pairs_sent.append(text_pairs)
        
    pairs.append(text_pairs_sent)
```

```python
data = list(zip(texts, pairs))
```
