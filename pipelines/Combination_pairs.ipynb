```python
from itertools import combinations
with open('rijeci5.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()
    test_list = [line.strip() for line in lines]
print('the original list :' + str(test_list))
res = list(combinations(test_list,2))
print ('all possible pairs:'+str(res))

for pair in res:
    print('({}, {})'.format(pair[0], pair[1]))
    
#tri su način zapisa liste. nakon ovog imamo stupac koji se onda učitava u SpyCy_arh_pairs.
#redovi se koriste za računanje, za Fischerov program.
```
