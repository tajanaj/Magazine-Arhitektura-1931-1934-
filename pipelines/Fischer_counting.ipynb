```python
import pandas as pd

# pročitajte tekstualnu datoteku, iskopiraj iz gornjeg prozora
with open('podaci36.txt', 'r', encoding='utf-8') as f:
    linije = f.readlines()

# pretvorite linije u listu rječnika
podaci = []
for linija in linije:
    riječi = linija.strip().split()
    podaci.append({'Prvi pojam': riječi[0], 'Drugi pojam': riječi[1], 'Broj': int(riječi[2])})

# pretvorite podatke u DataFrame
df = pd.DataFrame(podaci)

# spremite DataFrame u Excel datoteku
df.to_excel('tablica36.xlsx', index=False)
```
