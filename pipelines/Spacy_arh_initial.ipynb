```python
from collections import Counter
import spacy
import textacy
```

```python
nlp = spacy.load("hr_core_news_lg")
```

```python

with open ("rezultati/master.txt", mode='r', encoding='utf-8') as f:
    text_master = f.read().replace("\n", " ")
```

```python
doc = nlp(text_master)
```

```python
useless = ['ovaj', 'drugi', 'mnogo', 'taj', 'svaki', 'svoj', 'jedan', 'prema', 'koji', 'dva', 'prvi']
```

```python
lemma_words_master = []

for token in doc:
    if not token.is_stop and not token.is_punct and not token.is_space and token.lemma_ not in useless:
        lemma_words_master.append(token.lemma_)
        
print(lemma_words_master)
```

```python
word_freq = Counter(lemma_words_master)
master_words = word_freq.most_common(60)
master_keys = [k for k, value in word_freq.most_common(60)]
print(master_words)
#print(master_keys)
```

```python
import os
directory = "data"
texts = []

for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        f = os.path.join(directory, filename)
        texts.append(f)
```

```python
text_common = []
for text in texts:
    with open (text, mode='r', encoding='utf-8') as f:
        txt = f.read().replace("\n", " ")
    
    doc = nlp(txt)
    
    lemma_words = []
    for token in doc:
        if not token.is_stop and not token.is_punct and not token.is_space and token.text in master_keys: 
            if token.lemma_ not in useless:
                lemma_words.append(token.lemma_)
        
    #print(lemma_words)        
    word_freq = Counter(lemma_words)
    common_words = word_freq.most_common(10)
    text_common.append(common_words)
```

```python
data = list(zip(texts, text_common))
```
